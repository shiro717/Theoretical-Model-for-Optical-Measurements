```python
import numpy as np
import matplotlib.pyplot as plt

alpha = 7.29735e-3
n_s = 3.46

theta = np.linspace(-np.pi/2, np.pi/2, 1000)

sqrt_term = np.sqrt(n_s**2 - np.sin(theta)**2)

theta_F = np.arctan(2 * alpha * np.cos(theta)**2 / (sqrt_term + np.cos(theta)))
theta_K = np.arctan((4 * alpha * np.cos(theta)**2 * sqrt_term) / (n_s**2 - 1))

plt.figure(figsize=(8, 5))
plt.plot(theta, theta_F, label=r'$\theta_F$')
plt.plot(theta, theta_K, label=r'$\theta_K$')

plt.xlabel(r'$\theta$ (radians)')
plt.ylabel(r'$\theta_F, \theta_K$ (radians)')
plt.title(r'Faraday and Kerr Rotation Angles')
plt.legend()
plt.grid(True)

plt.show()
```

```python
E_z_t = - 2 * alpha * np.sin(2 * theta) * np.cos(theta) / ((sqrt_term + np.cos(theta))**2 + 4 * alpha**2 * (np.cos(theta))**2)
E_z_r = 2 * alpha * np.sin(2 * theta) * np.cos(theta) * sqrt_term / ((sqrt_term + np.cos(theta))**2 + 4 * alpha**2 * (np.cos(theta))**2)

plt.figure(figsize=(8, 5))
plt.plot(theta, E_z_t, label=r'$E_z^t/{E_x^i}$')
plt.plot(theta, E_z_r, label=r'$E_z^r/{E_x^i}$')

plt.xlabel(r'$\theta$ (radians)')
plt.ylabel(r'$E_z^t/{E_x^i}, E_z^r/{E_x^i}$')
plt.title(r'Ratio of $E_z^t$ ($E_z^r$) and $E_x^i$')
plt.legend()
plt.grid(True)

plt.show()
```
